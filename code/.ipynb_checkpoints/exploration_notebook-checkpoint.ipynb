```python
import os
import sys
import h5py  # To interface with the HDF5 binary data format.
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Path to the data.
data_path = "../data/Galaxy10_DECals.h5"

# To get the images and labels from file
with h5py.File(data_path, 'r') as F:
    images = np.array(F['images'])
    labels = np.array(F['ans'])
```

```python
print(np.unique(labels, return_counts=True))
```

```python
# To convert the labels to categorical 10 classes
labels = tf.keras.utils.to_categorical(labels, 10)

# To convert to desirable type
labels = labels.astype(np.float32)
images = images.astype(np.float32)
```

```python
print(labels.shape)
```

```python
print(images.shape)
```

```python
# Display some random selected images.
indices_to_inspect = range(0, 10000, 1000)

fig, ax = plt.subplots(1, 10)
fig.set_size_inches(12, 1.2)

for i, each_image in enumerate(indices_to_inspect):
    ax[i].imshow(images[each_image].astype('uint8'))
    ax[i].tick_params(left=False)
    ax[i].tick_params(bottom=False)
    ax[i].tick_params(labelleft=False)
    ax[i].tick_params(labelbottom=False)
    ax[i].set_xlabel(f"{each_image}")
    #ax[i].set_title(f"{labels[each_image]}")
```
